```python
df_1 = grouping_df[['index','tokenized_text','date']]
df_1.info()
```

```python
df_1['tokenized_text_05'] = [x.strip("[]").replace(',','').strip('"') for x in df_1['tokenized_text']]
df_1['tokenized_text_1'] = [ "".join(x) for x in df_1['tokenized_text_05']]
df_1
```

```python
df_2 = df_1.groupby('date')['tokenized_text_1'].apply(''.join)
df_3 = [x.split() for x in df_2]
(df_2)
df_3 = df_2
df_3 = pd.DataFrame(df_2)
df_3
df_3.to_csv('date_df_1.csv')
```
